```python
import csv
```

```python
lines = {}
with open("../data/RATP_to_check_with_osm.csv") as csvfile:
    csv_ = csv.DictReader(csvfile)
    for tt in csv_:
        current_line = tt["route_id"]
        if current_line in lines :
            lines[current_line].append(tt)
        else :
            lines[current_line] = []
            lines[current_line].append(tt)

d = {k:len(lines[k]) for k in lines}


for a_line in sorted(list(d.keys()), key=d.__getitem__, reverse=True) :
    print(lines[a_line][0]['name'])
    routepoints = [elem['stop_name'] for elem in lines[a_line]]
    routepoints_str = " ; ".join(routepoints)
    print("{} routepoints manquants : {}".format(len(lines[a_line]), routepoints_str))
    print(" https://ref-lignes-stif.5apps.com/line.html?osm_relation={}".format(lines[a_line][0]['line_id'].split(':')[-1]))
    print()
```
